### Hand Selected Training Data
The training data was carefully selected from a large spreadsheet to ensure that the classifier would be training on the right sentences and their corresponding labels

```python
train = [
    ("Is a credit check necessary for a new connection?","CREDIT CHECK RELATED"),
    ("Will you check my credit?","CREDIT CHECK RELATED"),
    ("A hard check will affect my credit score.","CREDIT CHECK RELATED"),
    ("Why is a credit check required?","CREDIT CHECK RELATED"),
  
    ("But unfortunately there is an outage in your area at the moment.","TECHNICIAN ISSUES"),
    ("Technician can not work when there is an outage.","TECHNICIAN ISSUES"),
    ("Why do I need a technician that is only available tomorrow?","TECHNICIAN ISSUES"),
    ("I was hoping that the technician could use my router, but he couldn't.","TECHNICIAN ISSUES"),

    ("The system stops me from getting service.","FREEZING/ ORDERING ISSUES"),
    ("The local office couldn't complete my order at this time.","FREEZING/ ORDERING ISSUES"),
    ("When I click Place Order for my new plan its stuck on  running white and gray buttons.","FREEZING/ ORDERING ISSUES")

      
    
    
]
```

### Training the Naive Bayes Classifier

```python
cl = NaiveBayesClassifier(train)
```

### Percent Accuracy
Testing the accuracy of the machines predictions with the new test data

```python
print("Percent Accuracy of Model:", round(cl.accuracy(test)*100,1), "%")
```

### Final Classification
Classification for one day's worth of data takes approximately 215 seconds to complete

```python
def execute():

    t0 = time.time()

    for i in range(your_matrix.shape[0]):
    
        print("Cell Number: ", i+1, " LABEL: ", cl.classify(your_matrix[i][0]), "  PROBABILITY: ", round(cl.prob_classify(your_matrix[i][0]).prob(cl.classify(your_matrix[i][0])),2))
    
    
    t1 = time.time()

    print("Time Elapsed: " + str(t1 - t0) + " s")
```

```python
execute()
```
