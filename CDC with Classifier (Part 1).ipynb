### Training the Naive Bayes Classifier

```python
cl = NaiveBayesClassifier(train)
```

### Percent Accuracy
Testing the accuracy of the machines predictions with the new test data

```python
print("Percent Accuracy of Model:", round(cl.accuracy(test)*100,1), "%")
```

### Final Classification
Classification for one day's worth of data takes approximately 100 seconds to complete

```python
def execute():

    t0 = time.time()

    for i in range(your_matrix.shape[0]):
    
        print("Cell Number: ", i+1, " LABEL: ", cl.classify(your_matrix[i][0]), "  PROBABILITY: ", round(cl.prob_classify(your_matrix[i][0]).prob(cl.classify(your_matrix[i][0])),2))
    
    
    t1 = time.time()

    print("Time Elapsed: " + str(t1 - t0) + " s")
```

```python
execute()
```
